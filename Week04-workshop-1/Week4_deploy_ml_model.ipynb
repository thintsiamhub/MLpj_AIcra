```python
from sklearn.linear_model import LinearRegression
import pickle
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt #dpi 600
import seaborn as sns
np.set_printoptions(suppress=True)

data = pd.read_csv('https://raw.githubusercontent.com/ekaratnida/Applied-machine-learning/master/Week04-workshop-1/data.txt')
print(data.head())
print(data.shape)
print(data.info())
print(data.describe())

#X = data[["area","rooms"]]
X = data.iloc[:,0:2]
print(X.head())

#y = data.price
#y = data.iloc[:,2]
y = data[["price"]]
print(y.head())

plt.figure(figsize = (15,4), dpi=600)
plt.subplot(121)
plt.scatter(X.iloc[:,0],y)
plt.xlabel("Size of house (X1)")
plt.ylabel("Price (Y)")

plt.subplot(122)
plt.scatter(X.iloc[:,1],y)
plt.xlabel("Number of Bedrooms (X2)")
plt.ylabel("Price (Y)")


#sns.heatmap(data)
"""Train model"""
print("Train")
lin_reg = LinearRegression()
lin_reg.fit(X, y) #training process
print(type(lin_reg.intercept_)) #theta0
print(type(lin_reg.coef_)) #theta1, theta2,...

"""Predict"""
print("Predict")
#X_test = np.array([[2000,6]])
X_test = pd.DataFrame(
    {
    "area":[2000],
    "rooms":[6]
    }
)
result = lin_reg.predict(X_test)
print(result)

"""Deploy"""
print("Deploy")
pickle.dump(lin_reg, open('model.sav','wb') )
from google.colab import files
files.download('model.sav')
```

##***Step 1 Import library***

```python
from sklearn.linear_model import LinearRegression
import pickle #type of file
import numpy as np #numpy จัดการmetrix
import pandas as pd #pandas จัดการตาราง
import os #os จัดการเซฟไฟล์
import matplotlib.pyplot as plt #dpi 600 #pyplot = plot graph
import seaborn as sns
np.set_printoptions(suppress=True) #set option ย่อตัวเลขหรือข้อความไม่ให้มัน print ออกมาเยอะ
```

##***Step 2 Load Data***

```python
data = pd.read_csv('https://raw.githubusercontent.com/ekaratnida/Applied-machine-learning/master/Week04-workshop-1/data.txt')
print(data.head()) #show the first 5 rows ถ้าอยาก run จากท้ายสุดให้เปลี่ยนเป็น tail
print(type(data.shape)) #shape = บอกขนาดของdata สูง*กว้าง, set{}ซ้ำไม่ได้, list[], tuple(), dict{'key':'value'}
print(data.info())
print(data.describe())
```

##***Step 3 Split Data(X and Y)***

```python
#X = data[["area","rooms"]] แบ่ง column
X = data.iloc[:,0:1]
print(X.head())

#y = data.price
#y = data.iloc[:,2]
y = data[["price"]]
print(y.head())
```

##***Step 4 Visualize Dataset***

```python
#paint.net: image tool (.tiff 600 dpi)
plt.figure(figsize = (19,5), dpi=600) #dpi = dot per ink
plt.subplot(131) #1=height 3=width 1=order of image
plt.scatter(X.iloc[:,0],y) #plot between Area:Price
plt.xlabel("Area (X1)")
plt.ylabel("Price (Y)")

plt.subplot(132)
plt.scatter(X.iloc[:,1],y) #Room vs Price
plt.xlabel("Room (X2)")
plt.ylabel("Price (Y)")

plt.subplot(133)
plt.scatter(X.iloc[:,1],X.iloc[:,0]) #Room vs Area
plt.xlabel("Room (X2)")
plt.ylabel("Area (X1)")
```

```python
#paint.net: image tool (.tiff 600 dpi)
plt.figure(figsize = (15,7), dpi=600) #dpi = dot per ink
plt.subplot(221)
plt.scatter(X.iloc[:,0],y) #plot between Area:Price
plt.xlabel("Area (X1)")
plt.ylabel("Price (Y)")

plt.subplot(222)
plt.scatter(X.iloc[:,1],y) #Room vs Price
plt.xlabel("Room (X2)")
plt.ylabel("Price (Y)")

plt.subplot(223)
plt.scatter(X.iloc[:,1],X.iloc[:,0]) #Room vs Area
plt.xlabel("Room (X2)")
plt.ylabel("Area (X1)")
```

##***Step 5 Train Model***

```python
#sns.heatmap(data)
"""Train model"""
print("Train")
lin_reg = LinearRegression()
lin_reg.fit(X, y) #training process
print(lin_reg.intercept_) #theta0
print(lin_reg.coef_) #theta1, theta2,...
# y = 89597.9 +139.2(area) - 8738.0 (room)
```

##***Step 6 Model usage(predict)***

```python
"""Predict"""
print("Predict")
#X_test = np.array([[2000]])
X_test = pd.DataFrame(
    {
    "area":[2000]
    }
)
result = lin_reg.predict(X_test)
print(result)
```

```python
"""Deploy"""
print("Deploy")
pickle.dump(lin_reg, open('model.sav','wb') )
from google.colab import files
files.download('model.sav')
```
